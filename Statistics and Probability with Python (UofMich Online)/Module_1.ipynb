<h2 color = 'blue'>Vector Dot Product And Its Graph</h2>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
brr = np.array([6,7,8,9,10])
dotpr = arr * brr
```

```python
def make_axes():
    plt.axvline(0, color = 'black', lw = 0.5)
    plt.axhline(0, color = 'black', lw = 0.5)
    #Scale the plot
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.gca().set_aspect('equal', adjustable = 'box')
    fig = plt.gca()
    return fig

empty_fig = make_axes()
```

```python
def add_vector_to_figure(figure, v, color = 'black', label = None):
    origin = np.array([0,0])
    figure.quiver(*origin, *v, scale = 1, scale_units='xy', angles='xy', color=color)
    if label is not None:
        figure.annotate(label, v, textcoords = "offset points", xytext=(10, 10), ha='center')
```

```python
c_vector = np.array([1,1])
d_vector = np.array([0.785,1.176])
e_vector = np.array([-1,1])
f_vector = np.array([-0.785,-1.176])
```

```python
empty_fig = make_axes()
```

```python
empty_fig = make_axes()
add_vector_to_figure(empty_fig, c_vector, label = "c")
```

```python
empty_fig = make_axes()
add_vector_to_figure(empty_fig, c_vector, color = 'blue', label = "c")
add_vector_to_figure(empty_fig, d_vector, color = 'red', label = "d")
add_vector_to_figure(empty_fig, e_vector, color = 'green', label = "e")
add_vector_to_figure(empty_fig, f_vector, color = 'yellow', label = "f")
```

```python
vecset = [c_vector, d_vector, e_vector, f_vector]
```

```python
print(vecset)
```

```python
sim2c = np.ones((4))
for i in range(len(vecset)):
    sim2c[i] = np.dot(c_vector, vecset[i])

```

```python
print(sim2c)
```

<h2>Application of Dot Product in Sentence Transformer Models</h2>

```python
from sentence_transformers import SentenceTransformer
```

```python
sentences = ['The Amazon rainforest is home to many species of animals.',
             'Neil armstrong walked on the moon in 1969.',
             'The forecast is cloudy with a chance of rain.',
             'Doctors recommed drinking plenty of water.',
             'The eagle soared high above the mountain peaks.']
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(sentences)
```

```python
embeddings.shape
```

```python
target = 'Consumption of disproportionately high amounts of protein may lead to difficulty in urination.'
```

```python
target_embedding = model.encode(target)
```

```python
target_embedding.shape
```

```python
matchdeg = embeddings * target_embedding
```

```python
matchdeg = np.dot(embeddings, target_embedding)
print(matchdeg)
```

```python
most_similar_sentence = sentences[np.argmax(matchdeg)]
print(most_similar_sentence)
```
